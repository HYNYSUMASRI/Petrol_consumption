```python
#step:3-create visualization
#barplot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x=df['Petrol_tax'].value_counts().index, y=df['Petrol_tax'].value_counts().values)
plt.title("Total Petrol_tax")
plt.xlabel("Petrol_tax")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Average_income'].value_counts().index, y=df['Average_income'].value_counts().values)
plt.xticks(rotation=60)
plt.title("Total Average_income")
plt.xlabel("Average_income")
plt.ylabel("Count")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Paved_Highways'], y=df['Population_Driver_licence(%)'], alpha=0.5)
plt.title("Paved_Highways vs Population_Driver_licence(%)")
plt.xlabel("Paved_Highways")
plt.ylabel("Population_Driver_licence(%)")
plt.show()
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Petrol_Consumption'], y=df['Petrol_tax'], alpha=0.5)
plt.title("Petrol_Consumption vs Petrol_tax")
plt.xlabel("Petrol_Consumption")
plt.ylabel("Petrol_tax")
plt.show()
```

```python
#step:4-data visualization
sns.histplot(df['Petrol_Consumption'], bins=10, kde=True)
plt.title("Petrol_Consumption")
plt.show()
```

```python
sns.boxplot(x=df['Petrol_Consumption'])
plt.title("Boxplot of Petrol_Consumption")
plt.show()
```

```python
#frequency polygon with height and width
plt.figure(figsize=(20,6))
sns.countplot(x=df['Petrol_Consumption'])
plt.title("Frequency Polygon of Petrol_Consumption")
plt.show()
```

```python
#step:5-feature enginnering
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head()
```

```python
df.Petrol_tax	=le.fit_transform(df.Petrol_tax)
df.Average_income = le.fit_transform(df.Average_income)
df.Paved_Highways = le.fit_transform(df.Paved_Highways)
df["Population_Driver_licence(%)"] = le.fit_transform(df["Population_Driver_licence(%)"])
df.Petrol_Consumption = le.fit_transform(df.Petrol_Consumption)
```

```python
df.head()
```

```python
X=df.drop(columns=['Petrol_Consumption'],axis=1)
X
```

```python
scaled_X=pd.DataFrame(scale.fit_transform(X),columns=X.columns)
scaled_X.head()
```
